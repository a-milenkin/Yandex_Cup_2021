```python
import pandas as pd
train_1 = pd.read_csv("reviews_ml.csv")
ratings=train_1.drop(['Unnamed: 0','ts','aspects','org_city','user_city'],axis=1)


ratings


```

```python
from surprise import Reader, Dataset

# to load dataset from pandas df, we need `load_fromm_df` method in surprise lib

ratings_dict = {'itemID': list(ratings.org_id),
                'userID': list(ratings.user_id),
                'rating': list(ratings.rating)}
df = pd.DataFrame(ratings_dict)

# A reader is still needed but only the rating_scale param is required.
# The Reader class is used to parse a file containing ratings.
reader = Reader(rating_scale=(4.0, 5.0))

# The columns must correspond to user id, item id and ratings (in that order).
data = Dataset.load_from_df(df[['userID', 'itemID', 'rating']], reader)
```

```python

# Split data into 5 folds

#data.split(n_folds=5)

from surprise import SVD,SVDpp
from surprise import NMF,BaselineOnly,NormalPredictor,KNNWithZScore,SlopeOne,CoClustering
from surprise.model_selection import cross_validate
from surprise.model_selection import GridSearchCV



# nmf
#algo = NMF()
#cross_validate(algo, data, measures=['RMSE'], cv=5, verbose=True)
#evaluate(algo, data, measures=['RMSE'])

param_grid = {'n_epochs': [100, 150],
              'lr_all':   [.002, .005],
              'reg_all':  [.6, .8]}

gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)

gs.fit(data)

# best RMSE score
print(gs.best_score['rmse'])

# combination of parameters that gave the best RMSE score
print(gs.best_params['rmse'])



```

```python
# svd
algo = SVD()
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
#evaluate(algo, data, measures=['RMSE'])

#BaselineOnly
algo = BaselineOnly()
cross_validate(algo, data, measures=['RMSE'], cv=5, verbose=True)


#NormalPredictor

algo = NormalPredictor()
cross_validate(algo, data, measures=['RMSE'], cv=5, verbose=True)

#SlopeOne

algo = SlopeOne()
cross_validate(algo, data, measures=['RMSE'], cv=5, verbose=True)

#SVDpp

algo = SVDpp()
cross_validate(algo, data, measures=['RMSE'], cv=5, verbose=True)

#CoClustering

algo = CoClustering()
cross_validate(algo, data, measures=['RMSE'], cv=5, verbose=True)
```
